```python
import numpy
import math
y1 = math.sin(1.0 / 5) * math.exp(1.0 / 10) + 5 * math.exp(-1.0 / 2)
```

```python
print (y1)
```

```python
y2 = math.sin(15.0 / 5) * math.exp(15.0 / 10) + 5 * math.exp(-15.0 / 2)
```

```python
print (y2)
```

```python
import numpy as np
a = np.array([[1, 1], [15, 1]])
b = np.array([y1, y2])
x = np.linalg.solve(a, b)
```

```python
print(x)
```

```python
import matplotlib
%matplotlib inline
from matplotlib import pylab as plt
x1 = np.arange(1, 15, 0.1)
y=x[0]*x1+x[1]
plt.plot(x1,y)
plt.show()
```

```python
y3 = math.sin(8.0 / 5) * math.exp(8.0 / 10) + 5 * math.exp(-8.0 / 2)
```

```python
a = np.array([[1, 1, 1], [225, 15, 1], [64, 8, 1]])
b = np.array([y1, y2, y3])
x = np.linalg.solve(a, b)
```

```python
print(x)
```

```python
x1 = np.arange(1, 15, 0.1)
y=x[0]*x1*x1+x[1]*x1+x[2]
plt.plot(x1,y)
plt.show()
```
